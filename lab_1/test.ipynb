```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#unit-step function
t=range(-400,400,1)
x=[i for i in t]
y=[0 if j<=0 else 1 for j in x]
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#unit-impulse function
y=[1 if j==0 else 0 for j in x]
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#ramp function
y=[0 if j<0 else j for j in x]
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
#periodic sinusoidal sequence
x = np.linspace(-np.pi*4, np.pi*4, 201)
y=np.sin(x)
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#periodic rectangular function
x=[i for i in range(-500,500)]
y=[]
i=x[0]
while i<=x[-1]:
    if i%150==0:
        for j in range(100):
            y.append(1)
        i+=100
    else:
        for j in range(50):
            y.append(0)
        i+=50
plt.grid()
plt.plot(x,y[:1000])
plt.show()
```

```python
#Asymmetric sawtooth waveform
def asymmetric_sawtooth(t, period=1.0, amplitude=1.0, rise_time=0.3, fall_time=0.7):
    t = np.mod(t, period)
    rise_duration = rise_time * period
    fall_duration = fall_time * period
    signal = np.where((t >= 0) & (t < rise_duration), amplitude * t / rise_duration, 0)
    signal += np.where((t >= rise_duration) & (t < period), amplitude * (1 - (t - rise_duration) / fall_duration), 0)
    return signal
     
t = np.linspace(0, 2, 1000)

period = 1.0
amplitude = 1.0
rise_time = 0.3
fall_time = 0.7
asymmetric_sawtooth_waveform = asymmetric_sawtooth(t, period, amplitude, rise_time, fall_time)

plt.figure(figsize=(10, 4))
plt.plot(t, asymmetric_sawtooth_waveform, label='Asymmetric Sawtooth Waveform')
plt.title('Asymmetric Sawtooth Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
     
```

```python
#Periodic Gaussian pulse
def periodic_gaussian_pulse(t, period=1.0, amplitude=1.0, mean=0.0, std_dev=0.1):
    t = np.mod(t, period)
    signal = amplitude * np.exp(-(t - mean)**2 / (2 * std_dev**2))
    return signal

t = np.linspace(0, 2, 1000)

period = 1.0
amplitude = 1.0
mean = 0.5
std_dev = 0.1
periodic_gaussian_pulse_waveform = periodic_gaussian_pulse(t, period, amplitude, mean, std_dev)

plt.figure(figsize=(10, 4))
plt.plot(t, periodic_gaussian_pulse_waveform, label='Periodic Gaussian Pulse')
plt.title('Periodic Gaussian Pulse')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#staircase function
x=[i for i in range(10)]
plt.stairs(x)

plt.show()
```

```python
t=np.linspace(0,10,1000)
def stair_case(t,step_points):
    y=np.zeros_like(t)
    for step_point in step_points:
        y+=(t>=step_point)
    return y

step_points=[1,2,3,4,5,6,7,8,9]
y=stair_case(t,step_points)
plt.plot(t,y)
plt.show()
```
