```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(t):
    return np.sin(t)
def y(t):
    return np.cos(t)

t=np.arange(-20,20,1)
```

```python
#x(t)
xt = x(t)
plt.stem(t,xt)
plt.show()
```

```python
#y(t)
yt = [y(i) for i in t]
plt.stem(t,yt)
plt.show()
```

```python
xt,yt=np.array(xt),np.array(yt)
convt = np.convolve(xt,yt,'same')
plt.stem(t,convt)
plt.show()
```

```python
#h(t)
def h(a):
    return np.tan(a)

ht = [h(i) for i in t]
plt.stem(t,ht)
plt.show()
```

```python
#question-2
ht=np.array(ht)
htyt=np.convolve(yt,ht,'same')
ytht=np.convolve(ht,yt,'same')

plt.stem(t,htyt)
plt.show()
```

```python

plt.stem(t,ytht)
plt.show()
```

```python
htyt=np.array(htyt)
ytht=np.array(ytht)
print(htyt-ytht)
```

```python
#question-3
def u(a):
    if a<0:
        return 0
    else:
        return 1

def delta(a):
    if a==0:
        return 1
    else:
        return 0

ut=[u(i) for i in t]
dt=[delta(i) for i in t]

plt.stem(t,ut)
plt.show()
```

```python
plt.stem(t,dt)
plt.show()
```

```python
def x(n):
    return 3*delta(n+2)-delta(n-1)+2*delta(n-3)

def h(n):
    return u(n+4)-u(n-3)

t=np.arange(-50,51)

xt=[x(i) for i in t]
ht=[h(i) for i in t]

plt.stem(t,xt)
plt.show()
```

```python
plt.stem(t,ht)
plt.show()
```

```python
convt = np.convolve(xt,ht,'same')
plt.stem(t,convt)
plt.show()
```

```python
#question-4
n = np.arange(-2, 3)

x = np.sin(2 * np.pi * n / len(n))

x = x + np.random.normal(0, 0.1, len(n))

h1 = np.ones(3) / 3
h2 = np.ones(3) / 7
h3 = np.ones(11) / 11

y1 = np.convolve(x, h1, 'same')
y2 = np.convolve(x, h2, 'same')
y3 = np.convolve(x, h3[:len(n)], 'same')

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.stem(n, x)
plt.title('Signal x(n)')

plt.subplot(4, 1, 2)
plt.stem(n, y1)
plt.title('Response y1(n)')

plt.subplot(4, 1, 3)
plt.stem(n, y2)
plt.title('Response y2(n)')

plt.subplot(4, 1, 4)
plt.stem(n, y3)
plt.title('Response y3(n)')

plt.tight_layout()
plt.show()
```
