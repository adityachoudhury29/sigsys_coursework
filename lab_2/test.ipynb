```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(t):
    if 0 <= t < 5:
        return 1
    elif 5 <= t < 8:
        return 2
    elif 8 <= t < 12:
        return 5
    else:
        return 0
def y(t):
    if 0 <= t < 7:
        return 2
    elif 7 <= t < 10:
        return 0
    elif 10 <= t < 15:
        return 7
    else:
        return 0

t=range(-5,16)
```

```python
#plotting x(t)
xt = [x(i) for i in t]
plt.plot(t,xt)
plt.grid()
plt.show()
```

```python
#plotting y(t)
yt = [y(i) for i in t]
plt.plot(t,yt)
plt.grid()
plt.show()
```

```python
#x(t)+y(t)
xt=np.array(xt)
yt=np.array(yt)
sumt=xt+yt
plt.plot(t,sumt)
plt.grid()
plt.show()
```

```python
#x(t)-y(t)
difft=xt-yt
plt.plot(t,difft)
plt.grid()
plt.show()
```

```python
#x(t)*y(t)
prodt=xt*yt
plt.plot(t,prodt)
plt.grid()
plt.show()
```

```python
#x(t)/2 + y(t)/3
funt=xt/2+yt/3
plt.plot(t,funt)
plt.grid()
plt.show()
```

```python
#x(-t)
t=range(-15,6)
minusxt=[x(-1*i) for i in t]
plt.plot(t,minusxt)
plt.grid()
plt.show()
```

```python
#y(-t)
t=range(-15,6)
minusyt=[y(-1*i) for i in t]
plt.plot(t,minusyt)
plt.grid()
plt.show()
```

```python
#x(2t)
t=range(-5,20)
x2t=[x(2*i) for i in t]
plt.plot(t,x2t)
plt.grid()
plt.show()
```

```python
#x(-2t+5)
t=range(-5,20)
x2t=[x(-2*i+5) for i in t]
plt.plot(t,x2t)
plt.grid()
plt.show()
```

```python
#x(0.5t-5)
t=range(-10,40)
ft=[x(0.5*i-5) for i in t]
plt.plot(t,ft)
plt.grid()
plt.show()
```

```python
#x(-0.5t-5)
t=range(-40,10)
ft=[x(-0.5*i-5) for i in t]
plt.plot(t,ft)
plt.grid()
plt.show()
```
