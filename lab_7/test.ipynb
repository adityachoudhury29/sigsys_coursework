```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,20,100)
n=np.linspace(0,20,20)
def x(t):
    return np.sin(t)
```

```python
#1
y1=t*x(t)
y2=np.exp(x(t))
y3=x(t+10)+(x(t))**2
y4=np.sin(x(n))
y5=x(n+1)+x(n)+x(n-1)

plt.subplot(2,3,1)
plt.plot(t,y1)
plt.subplot(2,3,2)
plt.plot(t,y2)
plt.subplot(2,3,3)
plt.plot(t,y3)
plt.subplot(2,3,4)
plt.plot(n,y4)
plt.subplot(2,3,5)
plt.plot(n,y5)
plt.show()
```
